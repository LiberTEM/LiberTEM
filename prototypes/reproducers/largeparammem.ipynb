```python
import numpy as np

from libertem.api import Context
from libertem.executor.inline import InlineJobExecutor
from libertem.udf.base import UDF
```

```python
from libertem import __version__
print(__version__)
```

```python
ctx = Context()
```

```python
inline_ctx = Context(executor=InlineJobExecutor())
```

```python
class NoOpUDF(UDF):
    def process_partition(self, partition):
        pass
    
    def get_result_buffers(self):
        return {
        }
    
    def get_backends(self):
        return ('numpy', 'cupy')
    
    def merge(self, dest, src):
        pass
```

```python
filename = '/cachedata/weber/tmpsparse.raw'
```

```python
# WARNING only do this on Linux or a machine with lots of RAM!
# This creates a sparse file on Linux if the file system supports it
# np.memmap(filename, shape=(2048 * 4096, 64, 64), dtype=np.float32, mode='w+')
```

```python
ds = ctx.load(
    'RAW',
    path=filename, 
    nav_shape=(2048*4096, ),
    sig_shape=(64, 64),
    dtype=np.float32,
)
```

```python
A = np.zeros((8*2048, 4096))
```

```python
inline_ctx.run_udf(dataset=ds, udf=NoOpUDF(A=A), progress=True, backends=('numpy',))
```

```python
ctx.run_udf(dataset=ds, udf=NoOpUDF(A=A), backends=('numpy',), progress=True)
```
